## ElasticNet

```python
from sklearn.linear_model import ElasticNet
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

First remove any with too many zeroes.

```python
sam = spread_adult_merged.copy()
```

```python
sam.head()
```

```python
# sam = spread_adult_merged.copy()
for code, name in metric_id:
    v = sam[code].value_counts()
    if 0 in v.index and v[0] > 0.6*sum(v):
        sam.drop(columns=code, inplace=True)
        
sam.columns
```

```python
X = sam.iloc[:,1:24]    # set X - independent vars
y = sam['Adult_Exp_PMPM']
```

```python
train_x, test_x, train_y, test_y = train_test_split(X, y, test_size=0.3, random_state=1)
print(train_x.shape)
print(test_x.shape)
print(train_y.shape)
print(test_y.shape)
```

### Lasso

```python
lm_lasso = Lasso(alpha=1, random_state=0)
lm_lasso.fit(train_x, train_y)
```

```python
from pprint import pprint
pprint(list(zip(train_x.columns, lm_lasso.coef_)))
```

```python
plt.figure(figsize=(15,10))
ft_imp_lm = pd.Series(lm_lasso.coef_, index = train_x.columns)
ft_imp_lm.plot(kind='barh')
plt.show();
```

```python
# R^2 score for the model
np.round(lm_lasso.score(train_x, train_y)*100,3)
```

```python
predict_test_lasso = lm_lasso.predict(test_x)
```

```python
print("Lasso MSE: ", metrics.mean_squared_error(test_y, predict_test_lasso))
```

### Ridge

```python
lm_ridge = Ridge(alpha=1, normalize=True, random_state=1)
lm_ridge.fit(train_x, train_y)
```

```python
from pprint import pprint
pprint(list(zip(train_x.columns, lm_ridge.coef_)))
```

```python
plt.figure(figsize=(15,10))
ft_imp_lm = pd.Series(lm_ridge.coef_, index=train_x.columns)
ft_imp_lm.plot(kind='barh')
plt.show();
```

```python
# R^2 score for the model
np.round(lm_ridge.score(train_x, train_y)*100,3)
```

```python
predict_test_ridge = lm_ridge.predict(test_x)
```

```python
print("Ridge MSE: ", metrics.mean_squared_error(test_y, predict_test_ridge))
```

### ElasticNet

```python
lm_elastic = ElasticNet(alpha=0.001, l1_ratio=0.5, normalize=True, random_state=1)
lm_elastic.fit(train_x, train_y)
```

```python
from pprint import pprint
pprint(list(zip(train_x.columns, lm_elastic.coef_)))
```

```python
plt.figure(figsize=(15,10))
ft_imp_lm = pd.Series(lm_elastic.coef_, index=train_x.columns)
ft_imp_lm.plot(kind='barh')
plt.show();
```

```python
# R^2 score for the model
np.round(lm_elastic.score(train_x, train_y)*100,3)
```

```python
predict_test_elastic = lm_elastic.predict(test_x)
```

```python
print("ElasticNet MSE: ", metrics.mean_squared_error(test_y, predict_test_elastic))
```

## PCA

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
pca.fit(X)
```

```python
print(pca.explained_variance_ratio_)
```

```python
two_d_data = pca.transform(X)
```

```python
two_d_data.shape
```

```python
# y_axis_zeros = np.zeros(len(one_d_data))
plt.scatter(two_d_data[:,0], two_d_data[:,1])
```

```python
# look at what features are being assigned high weight
print(pca.components_[0])
print(pca.components_[1])
```

```python
# get in abs value
abs_val_two_d_weights_component_1 = np.abs(pca.components_[0])
abs_val_two_d_weights_component_2 = np.abs(pca.components_[1])

ranking_1 = np.argsort(-abs_val_two_d_weights_component_1)
ranking_2 = np.argsort(-abs_val_two_d_weights_component_2)

# now print out the metrics with highest to lowest absolute weight
print(f'First Component, which explains {np.round(pca.explained_variance_ratio_[0]*100, decimals=2)}% of the variance')
print("="*80)
for r in ranking_1:
    print(metric_id[r], ':', pca.components_[0][r])

print('\n\n')
print(f'Second Component, which explains {np.round(pca.explained_variance_ratio_[1]*100, decimals=2)}% of the variance')
print("="*80)
for r in ranking_2:
    print(metric_id[r], ':', pca.components_[1][r])
```
